### Issue #7:

#### Define
Create predict_breed, value is follow sequence equal p1 if p1_dog is True, if not it equal p2 if p2_dog is True...<br>
And delete redundant columns not use

#### Code

```python
image_predictions_clean.head()
```

```python
def most_predict (row):
    if row['p1_dog']:
        return row['p1']
    elif row['p2_dog']:
        return row['p2']
    elif row['p3_dog']:
        return row['p3']
    else:
        return 'None'

image_predictions_clean['predict_breed'] = image_predictions_clean.apply(most_predict, axis=1)
```

### Tidiness issue #2:

#### Define
Convert the dog stage into one column instead of the multiple columns.<br>
And delete redundant column when done convert

#### Code

```python
twitter_archive_clean['stage_concat'] = twitter_archive_clean.doggo.str.capitalize() + ', ' + twitter_archive_clean.floofer.str.capitalize() + ', ' + twitter_archive_clean.pupper.str.capitalize() + ', ' + twitter_archive_clean.puppo.str.capitalize()
```

```python
twitter_archive_clean.stage_concat.value_counts()
```

```python
def transform_stage(data):
    if data['stage_concat'].count('None') == 2:       
        return data['stage_concat'].replace('None, ', '').replace('None', '')[:-2]
    else:
        if data['stage_concat'].count('Doggo') == 1:
            return 'Doggo'
        elif data['stage_concat'].count('Floofer') == 1:
            return 'Floofer'
        elif data['stage_concat'].count('Pupper') == 1:
            return 'Pupper'
        elif data['stage_concat'].count('Puppo') == 1:
            return 'Puppo'
        else:
            return 'None'

twitter_archive_clean['dog_stage'] = twitter_archive_clean.apply(transform_stage, axis=1)
```

```python
twitter_archive_clean.drop(['stage_concat', 'doggo', 'floofer', 'pupper', 'puppo'], axis=1,inplace= True)
```

```python
twitter_archive_clean.dog_stage = twitter_archive_clean.dog_stage.astype('category')
```
