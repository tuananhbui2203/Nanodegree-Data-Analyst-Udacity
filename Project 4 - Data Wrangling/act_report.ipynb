# ANALYSIS AND VISUALIZATION

WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
twitter_archive_clean = pd.read_csv('twitter_archive_master.csv')
```

## 1.Most popular dog breed?

```python
breeds_top_10 = twitter_archive_clean.predict_breed.value_counts()[1:10]
breeds_top_10.plot.bar()
plt.title('Most popular Dog breeds')
plt.xlabel('Breeds')
plt.ylabel('Number of occurrence');
```

Labrador Retriever is the most popular dog breed in this event.

## 2. Most used Twitter Source?

```python
source = twitter_archive_clean['source'].value_counts()
source.plot.bar()
plt.title('Most used Twitter source')
plt.xlabel('Source')
plt.ylabel('Number of tweets');
```

Twitter for Iphone is the most popular source in this event. Beside is twitter web client and tweetdeck.

## 3. Top 5 breeds with a favorite count above 10000?

```python
twitter_archive_clean[twitter_archive_clean.favorite_count>10000].predict_breed.value_counts()
plt.pie(twitter_archive_clean[twitter_archive_clean.favorite_count>10000].predict_breed.value_counts().head(5));
plt.legend(['Golden Retriever','Labrador Retriever','Pembroke ','Chihuahua','Samoyed']);
plt.title('top 5 breeds with a favourite count above the 10000');
```

Golden Retriever, Labrador Retriever, Pembroke, Chihuahua, Samoyed is top 5 breeds with a favorite count above 10000.

## 4. Correlation between favorite and retweet?

```python
twitter_archive_clean.plot(x='favorite_count', y='retweet_count', kind='scatter', title = 'Favorite and Retweet');
```

```python
twitter_archive_clean['favorite_count'].corr(twitter_archive_clean['retweet_count'])
```

The above scatter plot shows there is a positive correlation between favorite and retweet and with the correlation coefficient ~ 0.9 it is a strong positive correlation
